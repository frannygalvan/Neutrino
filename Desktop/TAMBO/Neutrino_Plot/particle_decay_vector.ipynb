```julia
using LinearAlgebra
using Plots
using YAML
```

```julia
# Load the YAML file and inspect its structure
config = YAML.load_file("config2.yaml")

# Print the entire YAML content
println("YAML content:")
println(config)
```

```julia
# Extract the necessary values for the vector from the YAML file
zenith = 99.99732787664736
azimuth = 203.76835088101396
xpos = 5.82690291372798 * 1000  # Convert to meters
ypos = 2.651239329640523 * 1000  # Convert to meters
zpos = -0.7696777502523529 * 1000  # Convert to meters

# Display the extracted values
println("Zenith: ", zenith)
println("Azimuth: ", azimuth)
println("X Position: ", xpos)
println("Y Position: ", ypos)
println("Z Position: ", zpos)
```

```julia
# Convert zenith and azimuth from degrees to radians
zenith_rad = deg2rad(zenith)
azimuth_rad = deg2rad(azimuth)

# Convert spherical to Cartesian coordinates
vx = sin(zenith_rad) * cos(azimuth_rad)
vy = sin(zenith_rad) * sin(azimuth_rad)
vz = cos(zenith_rad)

# Display the Cartesian coordinates of the vector
println("Vector in Cartesian coordinates: (", vx, ", ", vy, ", ", vz, ")")
```

```julia
# Plot the vector by itself
scatter3d([xpos], [ypos], [zpos], xlabel="X (m)", ylabel="Y (m)", zlabel="Z (m)", 
          title="Decay Particle Vector", markersize=5, color=:red, label="Decay Vector")

# Add the vector
quiver!([xpos], [ypos], [zpos], quiver=([vx], [vy], [vz]), arrow=true, label="")

# Save the plot
# savefig("decay_particle_vector.png")
```
